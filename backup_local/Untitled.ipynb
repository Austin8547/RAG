```python
from langchain_core.documents import Document     #importin the document structure
```

```python


# Creating a document
doc = Document(
    page_content="The Q3 revenue exceeded expectations by 15%.",
    metadata={
        "source": "financial_report_2023.pdf",
        "page": 12,
        "author": "Finance Dept"
    }
)

print(doc.page_content) ## Output: The Q3 revenue exceeded expectations by 15%.

print(doc.metadata['source'])  # Output: financial_report_2023.pdf
```
