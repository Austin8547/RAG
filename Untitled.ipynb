```python
from nltk.stem import PorterStemmer

# Initialize the stemmer
porter = PorterStemmer()

words = ["running", "generously", "happily", "university", "universal"]

print(f"{'Original':<15} | {'Stemmed':<15}")
print("-" * 35)

for word in words:
    stemmed = porter.stem(word)
    print(f"{word:<15} | {stemmed:<15}")
```

```python
from nltk.stem import SnowballStemmer

# Initialize the stemmer for English
snowball = SnowballStemmer(language='english')

words = ["running", "generously", "happily", "university", "universal"]

print(f"{'Original':<15} | {'Stemmed':<15}")
print("-" * 35)

for word in words:
    stemmed = snowball.stem(word)
    print(f"{word:<15} | {stemmed:<15}")
```

```python
from nltk.stem import LancasterStemmer

# Initialize the stemmer
lancaster = LancasterStemmer()

words = ["running", "generously", "happily", "university", "universal"]

print(f"{'Original':<15} | {'Stemmed':<15}")
print("-" * 35)

for word in words:
    stemmed = lancaster.stem(word)
    print(f"{word:<15} | {stemmed:<15}")
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemm = WordNetLemmatizer()

words = ["running", "cars", "better", "mice", "studies", "meeting"]

lemmatized_words = [lemm.lemmatize(w, wordnet.VERB) for w in words]
print(lemmatized_words)
```
